I think we've decided that 200 is our best bet for step 1.



This is the set of equations that gets us there. y is expected profit of a trade, and x is the bid price. The term (320 - x) represents the profit from a trade, and the rest of it represents the probability that a trade will happen. We have two equations because we need to account for the turtles' superstition that bars from trading at prices that fall between 200 and 250.

${y=\frac{\left(320\ -\ x\right)\left(\int_{160}^{x}1dp\right)}{\left(\int_{160}^{200}1dx\ +\ \int_{250}^{320}1dx\right)}\ \left\{160\ \le x\le200\right\}}$

${y=\frac{\left(320\ -\ x\right)\left(\int_{160}^{200}1dp\ -\ \int_{250}^{x}1dp\right)}{\left(\int_{160}^{200}1dx\ +\ \int_{250}^{320}1dx\right)}\ \left\{250\ \le x\le320\right\}}$

The maximum of this set of equations happens when x = 200.
