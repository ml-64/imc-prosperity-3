```python
# PAIR TRADING
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df = pd.read_csv("../Tutorial Round/Tutorial Log Files/376b5875-5a54-49cc-a478-f52105bad8c1.csv")

rr_df = df[df["product"] == "RAINFOREST_RESIN"]
kelp_df = df[df["product"] == "KELP"]

# plt.plot(rr_df["timestamp"].values, rr_df["mid_price"].values)
# plt.plot(kelp_df["timestamp"].values, kelp_df["mid_price"].values)

# Pair Trading test

import statsmodels.api as sm
from statsmodels.tsa.stattools import coint
from statsmodels.regression.rolling import RollingOLS

resin_vals = list(rr_df["mid_price"].values - 10000)
kelp_vals = list(kelp_df["mid_price"].values)
resin_d = list()
kelp_d = list()

ctr = 0
for x in range(len(resin_vals)-1):
    resin_d.append(resin_vals[ctr+1] - resin_vals[ctr])
    ctr += 1
resin_d.append(0.)

resin_vals = np.array(resin_vals).reshape((-1,1))
resin_d = np.array(resin_d).reshape((-1,1))

ctr = 0
for x in range(len(kelp_vals)-1):
    kelp_d.append(kelp_vals[ctr+1] - kelp_vals[ctr])
    ctr += 1
kelp_d.append(0.)

reg = LinearRegression().fit(resin_vals, kelp_d)
reg.score(resin_vals, kelp_d)
```

```python
# LINEAR REGRESSION ON PREVIOUS TRADES

df = pd.read_csv("../Tutorial Round/Tutorial Log Files/376b5875-5a54-49cc-a478-f52105bad8c1.csv")

kelp_df = df[df["product"] == "KELP"]
kelp_vals = list(kelp_df["mid_price"].values)

window = 6

def standard_error(y_true, y_pred, num_predictors):
    residuals = y_true - y_pred
    n = len(y_true)
    se = np.sqrt(np.sum(residuals**2) / (n - num_predictors))
    return se

X = list()

for x in range(len(kelp_vals) - window):
    to_app = list()
    for y in range(window):
        to_app.append(kelp_vals[x+y])
    X.append(to_app)

X = np.array(X)

# print(len(X))      

y = np.array(kelp_vals[window:])

# print(len(y))

reg = LinearRegression().fit(X, y)
# print(reg.score(X, y))
# print(reg.coef_)
# print(reg.intercept_)

# se = standard_error(y, reg.predict(X), 6)
# print(se)

plt.plot(kelp_df["timestamp"].values[window:], y)
plt.plot(kelp_df["timestamp"].values[window:], reg.predict(X))

```
