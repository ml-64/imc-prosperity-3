Step 3: Assemble a portfolio.

My hypothesis: I believe given how steeply transaction costs rise, we need to be very selective about our trades.

- Define the portfolio optimization problem

$\max_{p_1, ..., p_9} \sum_{i = 1}^{9} (10000r_i p_i - c(p_i))$ under the constraint $\sum_{i = 1}^{9} {|p_i| \leq 100} $

Where $r_i$ denotes the return anticipated for commodity $i$, $p_i$ is the percent with regard to a commodity $i$, and $c(x)$ is a cost function, relating percent to transaction cost.

- Define the returns on each commodity

5: 0.4
4: 0.25
3: 0.15
2: 0.1
1: 0.05

- Solve the optimization problem



```python
import numpy as np
from scipy.optimize import minimize

# Sentiments for each commodity
sentiments = {
    'Haystacks': -1,
    'RanchSauce': 1,
    'CactiNeedles': -5,
    'SolarPanels': -5,
    'RedFlags': 4,
    'VRMonocle': 4,
    'QuantumCoffee': -3,
    'Moonshine': 2,
    'StripedShirts': 3,
}

# Returns for each sentiment
returns_for_sentiments = [0, 0.05, 0.1, 0.15, 0.25, 0.4]

# Products
products = list(sentiments.keys())

# Returns for products
returns_for_products = np.array([sentiment/abs(sentiment)*returns_for_sentiments[abs(sentiment)] for _, sentiment in sentiments.items()])

# Cost function
def cost(percent):
  return 120*(percent**2)

def objective(p):
  costs = np.array([cost(x) for x in p])
  return -np.sum((10000*p)*returns_for_products - costs)

def objective_individual(p, return_for_product):
  c = cost(p)
  return (10000*p)*return_for_product - c

def abs_constraint(p):
  return 100 - np.sum(np.abs(p))

constraints = ({
    'type': 'ineq',
    'fun': abs_constraint
})

# Bounds
bounds = [(-100, 100) for _ in range(9)]

# Initial guess
p0 = np.zeros(9)

result = minimize(objective, p0, bounds=bounds, constraints=constraints)

# Output
if result.success:
  print("Optimal p:", result.x)
  print("Maximum objective value:", -result.fun)
else:
  print("Optimization failed:", result.message)

# Rounding option: Regular
p_rounded_regular = np.round(result.x).astype(int)

# Rounding option: Choosing the nearest percent which maximizes the objective function
p_rounded_max = np.empty(9)
floors = np.floor(result.x).astype(int)
ceils = np.ceil(result.x).astype(int)
for i, pair in enumerate(zip(floors, ceils)):
  p_rounded_max[i] = max(pair, key=lambda x: objective_individual(x, returns_for_products[i]))
print("Optimal p int:", p_rounded_max)



```
